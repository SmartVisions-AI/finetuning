```python
models = [
    "moslehGen/DeepSeek-R1-Distill-Llama-8B-500Steps",
    "moslehGen/DeepSeek-R1-Distill-Qwen-14B-500Steps"
]

chosen_model = models[0]
```

```python
from unsloth import FastLanguageModel
max_seq_length = 2048
dtype = None
load_in_4bit = True



model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = chosen_model,
    max_seq_length = max_seq_length,
    dtype = dtype,
    load_in_4bit = load_in_4bit,
)
```

```python
#https://docs.unsloth.ai/basics/running-and-saving-models/saving-to-gguf

#model.save_pretrained_gguf("dir", tokenizer, quantization_method = "q4_k_m")
```

```python
model.push_to_hub_gguf(chosen_model+"-GGUF-q4_k_m", tokenizer, quantization_method = "q4_k_m",token = "YOUR_HF_TOKEN")
```
